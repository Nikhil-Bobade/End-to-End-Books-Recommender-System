```python
number_rating=rating_with_books.groupby('Title')['Book-Rating'].count().reset_index() 
```

```python
number_rating.rename(columns={"Book-Rating":'num_of_rating'},inplace=True)
```

```python
final_rating=rating_with_books.merge(number_rating,on='Title')
```

```python
# letas take those books which got at least 50 rating of users
final_rating=final_rating[final_rating['num_of_rating']>=50]
```

```python
# Lets create a pivot table 


book_pivot=final_rating.pivot_table(columns='User-ID',index='Title',values='Book-Rating')
```

```python
book_pivot.head()
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

# Testing model

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(suggestion)):
        books=book_pivot.index[suggestion[i]]
        for j in books:
            if j==book_name:
                print(f"You Searched '{book_name}'\n")
                print(f"The Suggestion books are :\n ")
            else:
                print(j)    
```

```python
book_name="A Civil Action"

recommend_book(book_name)
```
